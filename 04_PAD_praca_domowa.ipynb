3. Wyczyść kolumnę tshirt_category normalizując nazwy produktów. Upewnij się, że dla każdej kategorii funkcjonuje jedna nazwa, np. zamiast white i wh jest tylko jedna wersja white. Skorzystaj z klasy string i metody apply;

```python
def nameNormalizer(name, replacer, stringToReplace):
  normalizedString = []
  for i in name.split(" "):
    if i == replacer:
      normalizedString.append(stringToReplace)
    else:
      normalizedString.append(i)

  return " ".join(normalizedString)

df.tshirt_category = df.tshirt_category.apply(lambda x: nameNormalizer(x, 'Wh', 'White'))
df.tshirt_category = df.tshirt_category.apply(lambda x: nameNormalizer(x, 'Tshirt', 'T-Shirt'))
df.tshirt_category = df.tshirt_category.apply(lambda x: nameNormalizer(x, 'Bl', 'Black'))
df.tshirt_category = df.tshirt_category.apply(lambda x: nameNormalizer(x, 'Tennis', 'T'))
df.tshirt_category.unique()
```

4. Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.

```python
def recognize(name, match):
  for i in name.split():
    if i in match:
      return i
  return 'unknown'

tshirt_colours = ("Black", "White")
tshirt_types = ("T-Shirt", "Shirt", "Hoodie")
tshirt_sizes = ("M", "F")

df["tshirt_type"] = df['tshirt_category'].apply(lambda x: recognize(x, tshirt_types))
df["tshirt_size"] = df['tshirt_category'].apply(lambda x: recognize(x, tshirt_sizes))
df["tshirt_colour"] = df['tshirt_category'].apply(lambda x: recognize(x, tshirt_colours))
df
```
