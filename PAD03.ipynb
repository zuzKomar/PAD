```python
import pandas as pd
import numpy as np

df = pd.read_csv('pad03/PAD_03_PD.csv', sep=';')
group_country=df.groupby('Country')
group_country.aggregate({'Country':['count']})
```

W nowej kolumnie ‘owned_goods’ dodaj zmienną, która będzie informowała o liczbie posiadanych dóbr (zliczyć ile dana osoba posiada w sumie dóbr).

```python
newcol=pd.DataFrame(df['owns_car']+df['owns_TV']+df['owns_house']+df['owns_Phone'], columns=['owned_goods'])
df = df.join(newcol)
df
```

Porównaj średnią posiadanych dóbr między kobietami i mężczyznami. Wynik zaokrąglij do 2 miejsc po przecinku.

```python
df.groupby('gender', as_index=False)['owned_goods'].mean().round(2)
```

Stwórz ramkę danych, która zawierać będzie w pierwszej kolumnie wartość zmiennej COUNTRY, w drugiej kolumnie, średnią posiadanych dóbr w danym kraju (Wynik zaokrąglij do 2 miejsc po przecink), a w trzeciej kolumnie minimum wieku.

```python
groupedCountry = df.groupby('Country', as_index=False)
groupedCountryMeanOwnedGoodsMinAge = groupedCountry.aggregate({'owned_goods':['mean'], 'Age': ['min']})
groupedCountryMeanOwnedGoodsMinAge['owned_goods'] = groupedCountryMeanOwnedGoodsMinAge['owned_goods'].round(2)
groupedCountryMeanOwnedGoodsMinAge
```
